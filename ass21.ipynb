Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
# Histogram:

# Use: Visualize the distribution of a single variable.
# Code Example:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Employee.csv")
sns.histplot(data['Age'])
plt.title('Histogram - Age')
plt.show()
```

```python
# Use: Explore relationships between multiple variables in a dataset.
# Code Example:
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data, hue='LeaveOrNot')
plt.title('Pair Plot - Employee taking leaves')
plt.show()
```

```python
# # Create a violin plot
# Use: Visualize the correlation matrix of variables in a dataset.
# Code Example:
sns.violinplot(x='ExperienceInCurrentDomain', y='PaymentTier', data=data)
plt.title('Violin Plot - Total Bill by Day')
plt.show()
```

```python
def change(data):
    if data=='Iris-setosa':
        return 1
    elif data=='Iris-virginica':
        return 2
    else:
        return 3
```

```python
data = pd.read_csv("iris.csv")
data['Species']=data['Species'].apply(change)
sns.heatmap(data.corr(), cmap='coolwarm', annot=True)
```

```python
sns.countplot(x='Species', data=data)
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title('Line Plot of "fmri" Dataset')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots side by side
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
def convert_month_to_number(month_name):
    month_dict = {
        'january': 1,
        'february': 2,
        'march': 3,
        'april': 4,
        'may': 5,
        'june': 6,
        'july': 7,
        'august': 8,
        'september': 9,
        'october': 10,
        'november': 11,
        'december': 12
    }
    lower_month_name = month_name.lower()
    if lower_month_name in month_dict:
        return month_dict[lower_month_name]
    else:
        return None

flights['month']=flights['month'].apply(convert_month_to_number)
flights.info()
sns.heatmap(flights)
```

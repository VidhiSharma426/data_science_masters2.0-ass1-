### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

The t-test and z-test are both statistical tests used to make inferences about population parameters based on sample data, but they have some key differences:

1. **Population Standard Deviation Known vs. Unknown:**
   - **Z-test:** It is used when the population standard deviation is known.
   - **T-test:** It is used when the population standard deviation is unknown and needs to be estimated from the sample.

2. **Sample Size:**
   - **Z-test:** It is more appropriate when the sample size is large (typically n > 30), as the distribution of the sample mean approaches a normal distribution due to the Central Limit Theorem.
   - **T-test:** It is often preferred when the sample size is small (less than 30) and the population standard deviation is unknown.

3. **Distribution:**
   - **Z-test:** Assumes a normal distribution of the population.
   - **T-test:** Assumes a t-distribution, which is wider and flatter than a normal distribution, accounting for the added uncertainty introduced by estimating the population standard deviation from the sample.

**Example Scenarios:**

- **Z-test:**
  Suppose you are analyzing the test scores of a population where the population standard deviation is known. For instance, you might be comparing the average test scores of a group of students to a known population mean.

- **T-test:**
  Imagine you are studying the effect of a new teaching method on a small sample of students (less than 30). Since you don't know the population standard deviation, you would use a t-test to compare the means of the two groups (e.g., control group vs. experimental group) and determine if there is a significant difference.

In summary, if you know the population standard deviation and have a sufficiently large sample, you might use a z-test. If the population standard deviation is unknown or you have a small sample, a t-test is more appropriate.

### Q2: Differentiate between one-tailed and two-tailed tests.

**One-Tailed Test:**

In a one-tailed test (also called a directional test), the critical region is defined on only one side of the sampling distribution. This means that you are interested in the effect in one direction, either an increase or a decrease. The choice of a one-tailed test is typically driven by a specific hypothesis or expectation about the direction of the effect.

- **Example:** Let's say you are testing a new drug's efficacy, and you expect it to improve the condition. Your hypotheses might be:
  - Null Hypothesis (H0): The drug has no effect.
  - Alternative Hypothesis (H1): The drug has a positive effect.

In this case, a one-tailed test would focus on the right side of the distribution, where values are higher than expected under the null hypothesis.

**Two-Tailed Test:**

In a two-tailed test (also called a non-directional test), the critical region is divided and defined on both sides of the sampling distribution. This allows you to detect any significant difference, whether it's an increase or a decrease. Two-tailed tests are generally used when there is no specific expectation about the direction of the effect or when you want to be sensitive to changes in either direction.

- **Example:** Consider a scenario where you are testing the average weight of a product. Your hypotheses might be:
  - Null Hypothesis (H0): The average weight is equal to a specified value.
  - Alternative Hypothesis (H1): The average weight is different from the specified value.

Here, a two-tailed test would consider both the left and right sides of the distribution, looking for deviations from the expected value in either direction.

**Summary:**

- **One-Tailed Test:** Used when there is a specific expectation about the direction of the effect.
- **Two-Tailed Test:** Used when you want to detect any significant difference, regardless of the direction.

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

In hypothesis testing, there are two types of errors that can occur: Type I error and Type II error.

1. **Type I Error (False Positive):**
   - **Definition:** Rejecting a true null hypothesis when it is actually true.
   - **Symbol:** Denoted by α (alpha).
   - **Probability:** The probability of committing a Type I error is the significance level (α), which is set by the researcher.

**Example Scenario for Type I Error:**
Suppose you are testing a new drug, and the null hypothesis (H0) is that the drug has no effect. The alternative hypothesis (H1) is that the drug is effective. If, based on your sample data, you reject the null hypothesis and conclude that the drug is effective when, in reality, it has no effect, you have committed a Type I error.

2. **Type II Error (False Negative):**
   - **Definition:** Failing to reject a false null hypothesis when it is actually false.
   - **Symbol:** Denoted by β (beta).
   - **Probability:** The probability of committing a Type II error is influenced by factors such as the sample size, effect size, and the chosen significance level.

**Example Scenario for Type II Error:**
Continuing with the drug example, let's say the null hypothesis (H0) is that the drug has no effect, and the alternative hypothesis (H1) is that the drug is effective. If, based on your sample data, you fail to reject the null hypothesis and conclude that the drug has no effect when, in reality, it is effective, you have committed a Type II error.

In summary:

- **Type I Error (False Positive):** Incorrectly rejecting a true null hypothesis.
  - Example: Concluding a new drug is effective when it has no effect.

- **Type II Error (False Negative):** Incorrectly failing to reject a false null hypothesis.
  - Example: Concluding a new drug has no effect when it is actually effective.

Researchers aim to minimize the risk of both types of errors but often face a trade-off: reducing the risk of one type of error may increase the risk of the other. The balance between Type I and Type II errors is managed through careful selection of significance levels, sample sizes, and statistical power.

### Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

Certainly! Let's go through a simple example using Bayes' Theorem.

**Example: Medical Testing**

Consider a diagnostic test for a rare disease. The disease affects 1% of the population. The diagnostic test has a true positive rate of 95% (probability of testing positive given that the person has the disease) and a false positive rate of 3% (probability of testing positive given that the person does not have the disease).

Let's say a person takes the test and receives a positive result. We want to use Bayes' Theorem to calculate the probability that the person actually has the disease.

Define the events:
- \( A \): Having the disease.
- \( B \): Testing positive.

Given probabilities:
- \( P(A) = 0.01 \) (prevalence of the disease in the population).
- \( P(B|A) = 0.95 \) (true positive rate).
- \( P(\neg A) = 0.99 \) (not having the disease, complement of \( A \)).
- \( P(B|\neg A) = 0.03 \) (false positive rate).

Now, we can use Bayes' Theorem to calculate \( P(A|B) \):

\[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

First, calculate the denominator \( P(B) \) using the law of total probability:

\[ P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A) \]

\[ P(B) = (0.95 \cdot 0.01) + (0.03 \cdot 0.99) \]

Now, substitute into the Bayes' Theorem formula:

\[ P(A|B) = \frac{0.95 \cdot 0.01}{(0.95 \cdot 0.01) + (0.03 \cdot 0.99)} \]

Calculate this expression to find the probability that the person actually has the disease given a positive test result.

After performing the calculations, you will obtain the updated probability that the person has the disease based on the new evidence (positive test result). This illustrates how Bayes' Theorem allows us to adjust our probability estimates in light of new information.

### Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

The margin of error (MOE) in a confidence interval is a measure of the uncertainty or precision associated with the estimate of a population parameter. It represents the range within which we expect the true population parameter to lie with a certain level of confidence. The formula for the margin of error in a confidence interval for the mean is typically:

\[ \text{Margin of Error} = Z \left( \frac{\sigma}{\sqrt{n}} \right) \]

Where:
- \( Z \) is the Z-score corresponding to the desired confidence level.
- \( \sigma \) is the population standard deviation.
- \( n \) is the sample size.

The margin of error is directly influenced by the standard deviation and inversely proportional to the square root of the sample size. Therefore, as the sample size increases, the margin of error tends to decrease, indicating increased precision in estimating the population parameter.

**Example: Effect of Sample Size on Margin of Error**

Suppose you want to estimate the average height of students in a school. You take two samples, one with a small size (\( n_1 \)) and another with a larger size (\( n_2 \)). Let's assume that the population standard deviation (\( \sigma \)) is known.

For the small sample (\( n_1 \)), the margin of error is calculated as:

\[ \text{MOE}_1 = Z \left( \frac{\sigma}{\sqrt{n_1}} \right) \]

For the larger sample (\( n_2 \)), the margin of error is calculated as:

\[ \text{MOE}_2 = Z \left( \frac{\sigma}{\sqrt{n_2}} \right) \]

If the two samples are taken under the same conditions and the only difference is the sample size, you'll observe that \( \text{MOE}_2 < \text{MOE}_1 \). In other words, the larger sample size results in a smaller margin of error, reflecting increased precision in estimating the true population mean.

A smaller margin of error implies that the confidence interval is narrower, providing a more precise range for the population parameter estimate. Researchers often aim for smaller margins of error to enhance the precision of their estimates, and achieving this often involves increasing the sample size.

### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

The Z-score (also called the standard score) is a measure of how many standard deviations a particular data point is from the mean of a distribution. It is calculated using the formula:

\[ Z = \frac{X - \mu}{\sigma} \]

where:
- \( X \) is the individual data point,
- \( \mu \) is the population mean,
- \( \sigma \) is the population standard deviation.

In this case, you have:
- \( X = 75 \) (the data point),
- \( \mu = 70 \) (the population mean),
- \( \sigma = 5 \) (the population standard deviation).

Now, substitute these values into the formula:

\[ Z = \frac{75 - 70}{5} \]

\[ Z = \frac{5}{5} \]

\[ Z = 1 \]

So, the Z-score for the data point with a value of 75 is 1.

**Interpretation:**
A Z-score of 1 means that the data point is 1 standard deviation above the mean in the distribution. In the context of a normal distribution, which is a bell-shaped curve, a Z-score of 1 is associated with the area under the curve to the right of the data point. This area represents the percentage of data points in the distribution that are below the given data point.

Interpreting the Z-score:
- The data point of 75 is higher than the mean (70).
- Being 1 standard deviation above the mean is a relatively common occurrence in a normal distribution.

In practical terms, a Z-score helps you understand how unusual or typical a particular data point is within the context of the entire distribution. Positive Z-scores indicate values above the mean, while negative Z-scores indicate values below the mean.
